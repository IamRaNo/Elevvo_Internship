```python
agg_funcs = {
    "order_item_id": "last",          # last item id
    "product_id": "last",             # last product id
    "shipping_limit_date": "last",    # last shipping limit
    "price": "sum", 
    "seller_id": "last",                  # sum of price
    "freight_value": "sum"            # sum of freight
}
```

```python
order_items_grouped=order_items.groupby("order_id", as_index=False).agg(agg_funcs)
```
